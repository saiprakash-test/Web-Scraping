```python
# Importing required libraries
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import pandas as pd
```

# TOP 10 MEN TEAM RANKINGS

```python
url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
r=requests.get(url)
soup=BeautifulSoup(r.content,'html.parser')
team=[]
match=[]
rating=[]
for i in soup.find_all('span',attrs={'u-hide-phablet'}):
    team.append(i.text)
match.append(soup.find('td',attrs={'rankings-block__banner--matches'}).text)
rating.append(soup.find('td',attrs={'rankings-block__banner--points'}).text)
count=0
for i in list(soup.find_all('td',attrs={'table-body__cell u-center-text'}))[:18]:
    if count%2==0:
        match.append(i.text)
    else:
        rating.append(i.text)
    count+=1
df=pd.DataFrame({'Team':team[:10],
                 'Match':match,
                 'Points':rating})
df.to_csv('Top10 ODI Men Teams.csv', index = False)
df
```

# INTERNATIONAL FIXTURES

```python
# Opening the Chromedriver
driver = webdriver.Chrome("chromedriver.exe")
import time
time.sleep(5)
```

```python
# Opening the homepage of bcci website
driver.get("https://www.bcci.tv/")
# Clicking on drop down navigation bar On International
driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover']").click()
# Selecting the Fixtures from International
driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a").click()
```

```python
urls=[]
match_page=driver.find_elements_by_xpath("//div[@class='js-list']/a")
for i in match_page:
    urls.append(i.get_attribute('href'))
```

```python
Match=[]
Series=[]
Place=[]
Date=[]
Time=[]

for i in urls:#list of urls of match description page
    driver.get(i)
    # Scraping the title by passing xpath
    try:
        match=driver.find_element_by_xpath("//p[@class='mc-header-info__match-description']/span")
        Match.append(match.text)
    except NoSuchElementException:#handling no such element exception
        Match.append('No Match available')
        
     # Scraping the series by passing xpath   
    try:
        series=driver.find_element_by_xpath("//h1[@class='mc-header-info__title']/strong")
        Series.append(series.text)
    except NoSuchElementException:#handling no such element exception
        Series.append("No Series Available")
        
     # Scraping the place by passing xpath   
    try:
        place=driver.find_element_by_xpath("//h1[@class='mc-header-info__title']")
        Place.append(place.text.split()[-1])
    except NoSuchElementException:#handling no such element exception
        Place.append("No Place Available")
        
     #scarping date of match   
    try: 
        date=driver.find_element_by_xpath("//div[@class='mc-header-scorebox__datetime']/strong")
        Date.append(date.text)
    except NoSuchElementException:#handling no such element exception
        Date.append("No Date Available") 
        
     #scarping time of the match   
    try:
        time=driver.find_element_by_xpath("//div[@class='mc-header-scorebox__datetime']/span")
        Time.append(time.text)
    except NoSuchElementException:#handling no such element exception
        Time.append("No Time Available")
```

```python
#Saving all the scraped data into a dataframe named BCCI
BCCI=pd.DataFrame({})
BCCI['Match Title']=Match
BCCI['Series']=Series
BCCI['Place']=Place
BCCI['Date']=Date
BCCI['Time']=Time
# Viewing the Scraped data
BCCI
```

```python
BCCI.to_csv("International Fixture details.csv",index = False)
```
